```python
import numpy as np
import matplotlib.pyplot as plt 
from astropy import units as u 
from astropy import constants as c 
import h5py
import scipy as stats
```

```python
f=h5py.File('blast2012_fields_with_units.hdf5','r')
```

Heat sink for charge controllers.

```python
Heat_SinkCC1=f['T_HS_CC1']
Heat_SinkCC2=f['T_HS_CC2']
```

Create a time array for flight time.

```python
Toime=np.array(f['TIME'])
```

```python
plt.figure(dpi=300)
plt.title('Heat Sink Temperatures')
plt.plot(Toime,Heat_SinkCC1, color='red', label='CC 1')
plt.plot(Toime,Heat_SinkCC2, color='blue', label='CC 2')
plt.legend()
plt.xlabel('Hours Since Launch')
plt.ylabel('Degrees Celsius')
plt.show()
```

Temperature of batteries.

```python
Temp_Bat1=f['T_1_BAT']
Temp_Bat2=f['T_2_BAT']
```

```python
plt.figure(dpi=300)
plt.title('Battery Temperatures')
plt.plot(Toime,Temp_Bat1, color='red', label='Batt 1')
plt.plot(Toime,Temp_Bat2, color='blue', label='Batt 2')
plt.plot
plt.xlabel('Hours Since Launch')
plt.ylabel('Degrees Celsius')
plt.show()
```
